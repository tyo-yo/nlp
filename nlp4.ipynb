```python
import MeCab 
import numpy as np 
```

```python
tagger = MeCab.Tagger()

with open("/Users/ryomisawa/Library/Mobile Documents/com~apple~CloudDocs/nlp_tutorial/neko.txt") as f,\
     open('neko.txt.mecab', mode='w') as mecab: 
     mecab.write(tagger.parse(f.read()))
```

```python
def neko_read():
    with open('neko.txt.mecab') as mecab:
        morphemes = []
        for line in mecab:
            cols = line.split('\t')
            if(len(cols)<2):
                raise StopIteration
            cols_1 = cols[1].split(',')

            morpheme = {
                'surface': cols[0],
                'base': cols_1[6],
                'pos': cols_1[0],
                'pos1' : cols_1[1]
            }
            morphemes.append(morpheme)

            if cols_1[1] == '句点':
                yield morphemes
                morphemes = []
```

```python
lines = neko_read()
surface_verbs = set()
for line in lines:
    for morpheme in line:
        if morpheme['pos'] == '動詞':
            surface_verbs.add(morpheme['surface'])
print(surface_verbs)
```

```python
lines = neko_read()
verbs = set()
for line in lines:
    for morpheme in line:
        if morpheme['pos'] == '動詞':
            verbs.add(morpheme['base'])
print(verbs)
```

```python
lines = neko_read()
a_no_b = set()
for line in lines:
    if len(line) > 2:
        for i in range(1, len(line)):
            if line[i]['surface'] == 'の' \
                and line[i-1]['pos'] == '名詞' \
                and line[i+1]['pos'] == '名詞':
                a_no_b.add(line[i-1]['surface'] + 'の' + line[i+1]['surface'])

print(a_no_b)
```

```python
noun_series = []
for line in neko_read():
    nouns = []
    for morpheme in line:
        if morpheme['pos'] == '名詞':
            nouns.append(morpheme['surface'])
        else:
            if len(nouns) > 1:
                noun_series.append("".join(nouns))
            nouns = []
    
    if len(nouns)>1:
        noun_series.append("".join(nouns))

set_noun_series = set(noun_series)
print(sorted(set_noun_series, key=noun_series.index))
```

```python
from collections import Counter
words = Counter()
for line in neko_read():
    words.update([morpheme['surface'] for morpheme in line])
print(words)
```
